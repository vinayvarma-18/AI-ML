```python
df1=df.copy()
df1
```

```python
datam=pd.DataFrame(df1)
datam['Age'].fillna(df['Age'].mean(), inplace=True)
datam
```

```python
datam.isna().sum()
```

```python
datam['Age'].fillna(datam['Age'].median(), inplace=True)
datam
```

```python
datam.isna().sum()
```

```python
#If any missing values in categorical columns, impute it with mode
datam=pd.DataFrame(df1)
datam['Embarked'].fillna(datam['Embarked'].mode()[0], inplace=True)
datam['Cabin'].fillna(datam['Cabin'].mode()[0], inplace=True)
datam
```

```python
datam.isna().sum()
```

```python
#For the categorical columns apply different kinds of encoding techniques
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
datam.Sex=le.fit_transform(datam.Sex)
datam.Name=le.fit_transform(datam.Name)
datam.Ticket=le.fit_transform(datam.Ticket)
datam.Cabin=le.fit_transform(datam.Cabin)
datam.Embarked=le.fit_transform(datam.Embarked)
datam
```
