# 2.Mean = 70 and Standard Deviation = 10 ,Identify how much area can be found between 70 and 90 with help of Empirical Rule.

### The empirical rule is a statistical rule which states that for a normal distribution, almost all data will fall within three standard deviations of the mean. Broken down, the empirical rule shows that 68% will fall within the first standard deviation, 95% within the first two standard deviations,99.7% within the first three standard deviations.

- As we know the area between 50 and 90 is 95% (i.e between first 2 standard deviations),the area between 70 and 90 will be 47.5% which is half the area of 95%.

### Therefore required area is 47.5%
